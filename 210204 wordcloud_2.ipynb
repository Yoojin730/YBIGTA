```python
del_ = ['위','시나','대한','번','의','를', '문단','인조','적','대상', '번','내용', '를', '부분', '공식', '온몸','것',
        '나', '중','듯', '곡', '티', '점','에스','프랴','참조','및','일','굿', '발표','처음', '그냥','자','이','경우',
        '족','수', '롤로','닉']
```

```python
i = 0

while i <= len(del_) - 1:
    while del_[i] in nouns :
        nouns.remove(del_[i])
    i = i + 1 
```

```python
def my_color_func(*args, **kwargs):
    #print(args, kwargs)
    font_size = kwargs['font_size']
    if font_size > 50:
        return "rgb(75,137,220)"
    elif font_size > 20:
        return "rgb(19,7,235)"
    else:
        return "rgb(9,216,235)"
```

```python
cinwc = WordCloud(font_path = font_path,
                  background_color = "white",
                  collocations = False,
                 color_func = my_color_func)

cinwc.generate(cin_text2)
```

```python
cinwc.to_image()
```
