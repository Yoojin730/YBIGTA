```python
import requests
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
from bs4 import BeautifulSoup
```

```python
url = 'https://namu.wiki/w/%EC%8B%9C%EB%82%98%EB%AA%A8%EB%A1%A4'
resp = requests.get(url)

soup = BeautifulSoup(resp.text, 'html.parser')
soup
```

```python
resp.status_code
```

```python
paragraph = soup.find_all('div', class_="wiki-paragraph")
paragraph
```

```python
type(paragraph)
```

```python
all_paragraph = []
for item in paragraph:
    all_paragraph.append(item.get_text().strip())
```

```python
file = open('cin.txt', 'w', encoding = 'utf-8')

for paragraph in all_paragraph :
    file.write(paragraph + '\n')
file.close()
```
